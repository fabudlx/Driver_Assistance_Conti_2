```python
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas
x_tail = 0.5
y_tail = 0.1

x_head = 0.9
y_head = 0.6
dx = x_head - x_tail
dy = y_head - y_tail
```

```python
fig, axs = plt.subplots()

arrow = mpatches.Arrow(x_tail, y_tail, dx, dy, width=0.5, color='green')
axs.add_patch(arrow)
axs.set_axis_off()
fig.add_axes(axs)
```

```python
fig.canvas.draw()

# Now we can save it to a numpy array.
data = np.fromstring(fig.canvas.tostring_rgb(), dtype=np.uint8, sep='')
data = data.reshape(fig.canvas.get_width_height()[::-1] + (3,))
data = cv.cvtColor(data, cv.COLOR_RGB2RGBA)
```

```python
data = cv.resize(data, (0,0), fx=1.5, fy=1.5) 
cv.imshow('Arrow',data)
cv.waitKey(0)
```
